```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import LatLon23 as latlon
import re 
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("global_temp_per_day_per_city_sortred.csv")
data.head()
```

```python
for i, row in data.iterrows():
    latArr = re.findall(r"\d+(?:\.\d+)?", data.at[i, 'Latitude']) + re.findall(r"\D", data.at[i, 'Latitude'])
    lat = str(latArr[0]+' '+latArr[2])
    lonArr = re.findall(r"\d+(?:\.\d+)?", data.at[i, 'Longitude']) + re.findall(r"\D", data.at[i, 'Longitude'])
    lon = str(lonArr[0]+' '+lonArr[2])
    location_i = latlon.string2latlon(lat, lon, 'D% %H').to_string()
    data.at[i, 'Latitude'] = location_i[0]
    data.at[i, 'Longitude'] = location_i[1]
```

```python
dates_list = data['dt'].unique()
dates_list
```

```python
pd.DataFrame(dates_list).to_csv('dates_list.csv')
```

```python
for day in dates_list:
    day_to_save = data.loc[data['dt'] == day]
    day_to_save.to_csv('chunks/global_temp_per_city_{}.csv'.format(day), index=False, header=True)
```
